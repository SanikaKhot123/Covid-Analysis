Generally, when the data is considered for analysis; the first thing to do is ask questions 
In the data analysis my top 3 questions are -
1. Which states have maximum number of covid cases?
2. Which states faced maximum deaths ?
3. Which states have maximum cured cases?

To answer, this questions I have did required analysis with required python Libraries.

The data used is from -
* State level data comes from Ministry of Health & Family Welfare.
* Testing data and vaccination data comes from covid19india. 

```python
#import all important libraries
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import datetime as dt
import numpy as np
```

```python
covid = pd.read_csv("../input/covid19-in-india/covid_19_india.csv", parse_dates=['Date'], dayfirst=True)
```

```python
covid.head()
```

```python
#selecting only required columns and renaming it 
covid = covid[['Date', 'State/UnionTerritory', 'Cured', 'Deaths', 'Confirmed']]

covid.columns = ['date', 'state', 'cured', 'deaths', 'confirmed']
```

```python
covid.head()
```

```python
covid.tail()
```

```python
today = covid[covid.date == '2021-05-23'] #added a date for analysis
now = covid[covid.date == '2022-04-22']
```

```python
now.head()
```

```python
now.shape #No active cases today
```

```python
today.shape  
```

***States with Maximum confirmed cases ***

```python
#sorting data w.r.t. confirmed cases
max_confirmed_cases = today.sort_values(by='confirmed', ascending=False)
```

```python
max_confirmed_cases.head()
```

```python
top_states_confirmed = max_confirmed_cases[0:5]
```

```python
sns.set(rc={'figure.figsize':(15, 10)})
sns.barplot(x="state", y="confirmed", data=top_states_confirmed, hue="state")
plt.show()
```

***States with maximum deaths due to Covid-19***

```python
max_death_cases = today.sort_values(by = 'deaths', ascending=False)
```

```python
max_death_cases.head(10)
```

```python
top_states_death = max_death_cases[0:5]
```

```python
sns.set(rc={'figure.figsize':(15, 10)})
sns.barplot(x="state", y="deaths", data=top_states_death, hue="state")
plt.show()
```

***States with high number of cured cases***

```python
max_cured_cases = today.sort_values(by = 'cured', ascending = False)
```

```python
max_cured_cases.head()
```

```python
top_states_cured = max_cured_cases[0:5]
```
