```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("../input/covid19-in-india/covid_19_india.csv", parse_dates=['Date'], dayfirst=True)
```

```python
#selecting only required columns and renaming it 
data = data[['Date', 'State/UnionTerritory', 'Cured', 'Deaths', 'Confirmed']]

data.columns = ['date', 'state', 'cured', 'deaths', 'confirmed']
```

```python
kerala = data[data.state=="Kerala"]
```

```python
kerala.shape
```

```python
kerala.head()
```

```python
kerala.tail()
```

**Kerala's Confirmed Cases**

```python
sns.set(rc={'figure.figsize':(15, 10)})
sns.lineplot(x="date", y="confirmed", data=kerala, color="red")
plt.show()
```

**Kerala Covid Death**

```python
sns.set(rc={'figure.figsize':(15, 10)})
sns.lineplot(x="date", y="deaths", data=kerala, color="black")
plt.show()
```

**Cured Covid cases in Kerala**

```python
sns.set(rc={'figure.figsize':(15, 10)})
sns.lineplot(x="date", y="cured", data=kerala, color="green")
plt.show()
```
